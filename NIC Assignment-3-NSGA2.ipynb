```python
import math
import random
import numpy as np
import matplotlib.pyplot as plt
import numpy as np

#Mazimise
def TotalVolume(h, r):
    total_vol = math.pi * r * r * ( 2/3 * r + h + h/9)
    return total_vol

#Minimise
def TotalSurfaceArea(h, r):
    tot_area = math.pi * r * (2*r + 2*h + math.sqrt(r*r + (h*h)/9))
    return tot_area

#Both H and R will take up 10 bits. (derived from their ranges)
def encoding_h_r(h, r):
    h = int(np.round(((h-min_h)*100)*(1023/1000)))              
    r = int(np.round(((r-min_r)*100)*(1023/800)))  
    h = bin(h)[2:]
    r = bin(r)[2:]
    h = h if len(h)==10 else ('0'*(10-len(h)) + h) #Pad zeroes.
    r = r if len(r)==10 else ('0'*(10-len(r)) + r)
    return h+r #Combine both.                                         

def decoding_h_r(ch):
    h = ch[:10]
    h = int(h,2)
    h = (h/100)*(1000/1023) + min_h

    r = ch[10:]
    r = int(r,2)
    r = (r/100)*(800/1023) + min_r

    return h,r

#NSGA-II's non dominated sort
def non_dominated_sort(values1, values2):
    S=[[] for i in range(0,len(values1))]
    front = [[]]
    n=[0 for i in range(0,len(values1))]
    rank = [0 for i in range(0, len(values1))]

    for p in range(0,len(values1)):
        S[p]=[]
        n[p]=0
        for q in range(0, len(values1)):
            if (values1[p] > values1[q] and values2[p] < values2[q]) or (values1[p] >= values1[q] and values2[p] < values2[q]) or (values1[p] > values1[q] and values2[p] <= values2[q]):
                if q not in S[p]:
                    S[p].append(q)
            elif (values1[q] > values1[p] and values2[q] < values2[p]) or (values1[q] >= values1[p] and values2[q] < values2[p]) or (values1[q] > values1[p] and values2[q] <= values2[p]):
                n[p] = n[p] + 1
        if n[p]==0:
            rank[p] = 0
            if p not in front[0]:
                front[0].append(p)

    i = 0
    while(front[i] != []):
        Q=[]
        for p in front[i]:
            for q in S[p]:
                n[q] =n[q] - 1
                if( n[q]==0):
                    rank[q]=i+1
                    if q not in Q:
                        Q.append(q)
        i = i+1
        front.append(Q)

    del front[len(front)-1]
    return front

def crowding_distance(values1, values2, front):
    distance = [0 for i in range(0,len(front))]
    sorted1 = sorting_values(front, values1[:])
    sorted2 = sorting_values(front, values2[:])
    distance[0] = float('inf')
    distance[len(front) - 1] = float('inf')

    for k in range(1,len(front)-1):
        distance[k] = distance[k]+ (values1[sorted1[k+1]] - values2[sorted1[k-1]])/(max(values1)-min(values1))
    for k in range(1,len(front)-1):
        distance[k] = distance[k]+ (values1[sorted2[k+1]] - values2[sorted2[k-1]])/(max(values2)-min(values2))
    return distance

def sorting_values(list1, values):
    sorted_list = []

    while(len(sorted_list)!=len(list1)):
        if return_index(min(values),values) in list1:
            sorted_list.append(return_index(min(values),values))

        values[return_index(min(values),values)] = math.inf

    return sorted_list

def crossover(ind1,ind2):
    cr_prob = np.random.randint(1, len(ind1))

    result1 = ind1[:cr_prob] + ind2[cr_prob:]
    result2 = ind1[cr_prob:] + ind2[:cr_prob]

    return result1,result2

def mutation(ind, mut_pro = 0.01):
    ind = np.array(list(map(int,list(ind)))).astype('bool')
    flip_bits = np.random.rand(len(ind)) < mut_pro

    #Flipping bits
    ind[flip_bits] = ~ind[flip_bits]
    ind = ind.astype(int)
    ind = "".join(list(map(str,ind)))

    return ind

def return_index(a,list):
    for i in range(0,len(list)):
        if list[i] == a:
            return i
    return -1

filenames = []

Size_population = 20
min_r=2
max_r=10
min_h=5
max_h=15

max_iterations = 100

r = [min_r+(max_r-min_r)*random.random() for i in range(0,Size_population)]
h = [min_h+(max_h-min_h)*random.random() for i in range(0,Size_population)]

iteration=0
while(iteration<max_iterations):
    function1_values = [TotalVolume(h[i],r[i])for i in range(0,Size_population)]
    function2_values = [TotalSurfaceArea(h[i],r[i])for i in range(0,Size_population)]
    non_dominated_sorted_solution = non_dominated_sort(function1_values[:],function2_values[:])

    print("Best Front for Iteration #", iteration, "is:")
    for val in non_dominated_sorted_solution[0]:
        print('( Radius =',round(r[val],3),', Height =',round(h[val],3),')',end=" ")
    print("\n")
    
    crowding_distance_values=[]
    for i in range(0,len(non_dominated_sorted_solution)):
        crowding_distance_values.append(crowding_distance(function1_values[:],function2_values[:],non_dominated_sorted_solution[i][:]))
    h_2 = h[:]
    r_2 = r[:]

    #Generating offsprings
    while(len(h_2)!=2*Size_population):
        a1 = random.randint(0,Size_population-1)
        b1 = random.randint(0,Size_population-1)
        individual_1 = encoding_h_r(h[a1],r[a1])
        individual_2 = encoding_h_r(h[b1],r[b1])
        individual_1, individual_2 = crossover(individual_1, individual_2)
        individual_1 = mutation(individual_1)
        individual_2 = mutation(individual_2)

        h_, r_ = decoding_h_r(individual_1)        
        if (5<= h_ and h_ <= 15) and (2<= r_ and r_ <= 10):
          h_2.append(h_)
          r_2.append(r_)

        h_, r_ = decoding_h_r(individual_2)
        if (5<= h_ and h_ <= 15) and (2<= r_ and r_ <= 10):
          h_2.append(h_)
          r_2.append(r_)

    function1_values2 = [TotalVolume(h_2[i],r_2[i])for i in range(0,2*Size_population)]
    function2_values2 = [TotalSurfaceArea(h_2[i],r_2[i])for i in range(0,2*Size_population)]

    non_dominated_sorted_solution2 = non_dominated_sort(function1_values2[:],function2_values2[:])
    crowding_distance_values2=[]
    
    for i in range(0,len(non_dominated_sorted_solution2)):
        crowding_distance_values2.append(crowding_distance(function1_values2[:],function2_values2[:],non_dominated_sorted_solution2[i][:]))
    
    new_solution= []
    for i in range(0,len(non_dominated_sorted_solution2)):
        non_dominated_sorted_solution2_1 = [return_index(non_dominated_sorted_solution2[i][j],non_dominated_sorted_solution2[i] ) for j in range(0,len(non_dominated_sorted_solution2[i]))]
        front22 = sorting_values(non_dominated_sorted_solution2_1[:], crowding_distance_values2[i][:])
        front = [non_dominated_sorted_solution2[i][front22[j]] for j in range(0,len(non_dominated_sorted_solution2[i]))]
        front.reverse()

        for value in front:
            new_solution.append(value)
            if (len(new_solution) == Size_population):
                break
        if (len(new_solution) == Size_population):
            break

    h = [h_2[i] for i in new_solution]
    r = [r_2[i] for i in new_solution]

    iteration = iteration + 1
    
    function1 = [i for i in function1_values]
    function2 = [j for j in function2_values]

    plt.xlim(0, 7500)
    plt.ylim(0, 2000)
    plt.xlabel('Total Volume')
    plt.ylabel('Total Surface Area (proportional to cost)')
    plt.title("Iteration: "+str(iteration))
    plt.scatter(function1, function2, color='red')

    x_y = [[function1[i],function2[i]] for i in range(len(function2_values))]
    x_y = sorted(x_y)
    x_y_x = [x_y[i][0] for i in range(len(x_y))]
    x_y_y = [x_y[i][1] for i in range(len(x_y))]
    plt.plot(x_y_x, x_y_y, color = 'green')

    filename = f'{iteration}.png'
    filenames.append(filename)
    plt.savefig('images/'+filename)
    plt.show()
    if (iteration==max_iterations):
      print("Final Objective function values:")
      final = []
      for i in range(len(function1)):
        print(function1[i], function2[i])
        l=[]
        l.append(function1[i])
        l.append(function2[i])
        final.append(l)

      final_copy = final.copy()
      final.sort()

      print('\n')
      print("Final Optimal Design:")  
      print("Total Volume:", final[len(final)//2][0]," ", "Total Surface Area:", final[len(final)//2][1])
      for g in range(len(h)):
        if ( final_copy[g][0] == final[len(final)//2][0] and final_copy[g][1] == final[len(final)//2][1] ):
          print("Height:",h[g],"Radius:",r[g])
          break
      print("Extreme objective values:")  
      print("1. Total Volume:", final[0][0]," ", "Total Surface Area:", final[0][1])
      print("2. Total Volume:", final[len(final)-1][0]," ", "Total Surface Area:", final[len(final)-1][1])

```
